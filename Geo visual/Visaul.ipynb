```python
import pandas as pd
import folium
data=pd.read_csv('Florida_Subsidence_Incident_Reports.csv')
```

```python
data_100=data.iloc[0:100,:]
florida=folium.Map(location=[28.5383, -81.3792],zoom_start=7)
map=folium.map.FeatureGroup()
for i,j in zip(data_100.Y,data_100.X):
    map.add_child(folium.CircleMarker([i,j],radius=5,color='red',fill_color='red'))
florida.add_child(map)
```
