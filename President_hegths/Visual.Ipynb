```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('president_heights.csv')
```

```python
# I want to see destribution of height of presidents
wheit=np.array(data['height(cm)'])
min_w=wheit.min()
max_w=wheit.max()
mean_w=wheit.mean()
std_w=wheit.std()
print('Minimal height is: ',min_w,'cm','\nMaximal height is: ',max_w,'cm','\nMean of height is: ',round(mean_w, 2),'cm','\nStandard deviation of height is: ',round(std_w,2),'cm')
```

```python
plt.hist(wheit,bins=10, edgecolor='black', color='lightblue')
plt.title('Mean of height is around 182 cm')
plt.xlabel('Height(cm)')
plt.ylabel('Number of presidents')
plt.show()
```
